### Reshaping, stacking, squeezing, unsqueezing

```python
a = torch.arange(0,4).reshape(2,2)
b = torch.arange(4,8).reshape(2,2)
```

```python
a.reshape(4)
```

```python
a.reshape(4,1)
```

```python
a.reshape(1,4)
```

```python
#viewwwwwwwwww
t = a.view(4)  #a is actually of (2,2) shape
t               #t shares the same memory a does
```

```python
t[0] = 1 #since they share the same memory making the changes to t will reflect on a aswell
t,a
```

```python
#stackingggggggg
a_stacked = torch.stack([a,b])
a_stacked
```

```python
a_stacked = torch.stack([a,b], dim=1)
a_stacked
```

```python
a,b
```

```python
a_stacked = torch.stack([a,b], dim=2)
a_stacked
```

```python
a_stacked[0], a[0], b[0]
```

### Squeezeeeeeee
removes all single dims from target tensor

```python
a_reshaped = a.reshape(1,1,4)
```

```python
a_reshaped, a_reshaped.shape
```

```python
a_squeezed = a_reshaped.squeeze()
a_squeezed
```

### Unsuqeeze
adds a single dimension to a target tensor at a specific dim

```python
a_unsqueezed = a_squeezed.unsqueeze(dim = 0)
a_unsqueezed, a_unsqueezed.shape
```

```python
a_unsqueezed  = a_squeezed.unsqueeze(dim = 1)
a_unsqueezed, a_unsqueezed.shape
```

### Permute
rearrangment of dims, returns view of tensor

```python
x_og = torch.rand(size=(224,224,3)) # color channel
x_og.shape
```

```python
x_permuted = x_og.permute(2,0,1) # rearrange by giving which dimenison to put where
x_permuted.shape                   # now it's rearranged to iykyk
```

```python
x_og[0,0,0] = 1
```

```python
x_og[0,0,0], x_permuted[0,0,0] # seeee, it is a view of the original tensor
```
